# How to solve the problem

### Problem
The value returned by nl2vis "return VegaLite(query2vl.to_VegaLite(pred_query, self.data)), query2vl.to_VegaLite(pred_query, self.data)" in Sevi.py don't render a chart.

### Experiments
* This is becasue VegaLite's returned value doesn't directly display the chart.
* Tested rendering Vega_Lite Chart via directly using HTML. Including Vega_Lite Json to HTML templet and rendering them using IPython.display module. (Didn't work)
* Tested Altair. However it doesn't work for this project since we need to dynamically change spec based on speech input. Altair define the spec using code.
* Found an example of spec structure from https://github.com/vega/ipyvega. 
* Combined the ipyvega's spec structure and the returned json structure from nl2vis. Then render them using IPython.display.

### Step1
* Sevi.Py > nl2vis: Update original code from return VegaLite(query2vl.to_VegaLite(pred_query, self.data)), query2vl.to_VegaLite(pred_query, self.data) to a new one as code below.

```python
from IPython.display import display

# VegaLite  JSON   
vega_lite_spec_json = query2vl.to_VegaLite(pred_query, self.data)

# JSON  $schema description  
vega_lite_spec_json["$schema"] = "https://vega.github.io/schema/vega-lite/v4.json"
vega_lite_spec_json["description"] = "A simple bar chart with embedded data."

#      
display({
    "application/vnd.vegalite.v5+json": vega_lite_spec_json
}, raw=True)
```
